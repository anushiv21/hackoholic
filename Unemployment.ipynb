```python
#unemployment rate year
uyear=sheet1.iloc[:,0].values
uyear
```

```python
#unemployment rate 
urate=sheet1.iloc[:,1].values
urate
```

```python
#youth unemployment rate
yrate=sheet1.iloc[:,3].values
yrate
```

```python
#participation rate
prate=sheet1.iloc[:,5].values
prate
```

```python
import matplotlib.pyplot as plt
plt.plot(uyear,urate,color="g")#green color
plt.ylabel("Unemployment Rate")
plt.xlabel("Year from 1990 to 2020")
plt.title("Unemployment Rate vs Year")
plt.show
```

```python
plt.plot(uyear,urate,color="b",label="unemployment rate")#green color
plt.plot(uyear,yrate,color="g",label="youth unemployment rate")#orange color
plt.legend()
plt.grid(True)
plt.ylabel("Unemployment and Youth Unemployment Rate")
plt.xlabel("Year from 1990 to 2020")
plt.title("Unemployment and Youth Unemployment Rate vs Year")
plt.show
```

```python
plt.bar(uyear,urate)
plt.ylabel("Unemployment Rate")
plt.xlabel("Year from 1990 to 2020")
plt.title("Unemployment Rate vs Year")
plt.xticks(uyear,rotation=70)
plt.show()
```

```python
plt.bar(uyear,yrate,color="g")
plt.ylabel("Youth Unemployment Rate")
plt.xlabel("Year from 1990 to 2020")
plt.title("Youth Unemployment Rate vs Year")
plt.xticks(uyear,rotation=70)
plt.show()
```

```python
plt.bar(uyear,prate)
plt.ylabel("Particiaption Rate")
plt.xlabel("Year from 1990 to 2020")
plt.title("Participation Rate vs Year")
plt.xticks(uyear,rotation=70)
plt.show()
```

```python
plt.bar(uyear,urate,color="b",label="unemployment rate")#green color
plt.bar(uyear,yrate,color="g",label="youth unemployment rate")#orange color
plt.legend()
plt.ylabel("Unemployment and Youth Unemployment Rate")
plt.xlabel("Year from 1990 to 2020")
plt.title("Unemployment and Youth Unemployment Rate vs Year")
plt.xticks(uyear,rotation=70)
plt.show()
print("here youth unemployment rate is completely overlapping unemployment rate.")
```

```python
#youth unemployment rate has completely overlapped unemployment rate in previous case so printed side by side.
ind=np.arange(len(uyear))
print(ind)
width=0.3
plt.bar(ind,urate,width,color="b",label="unemployment rate")#green color
plt.bar(ind+width,yrate,width,color="g",label="youth unemployment rate")#orange color
plt.legend()
plt.ylabel("Unemployment and Youth Unemployment Rate")
plt.xlabel("Year from 1990 to 2020")
plt.title("Unemployment and Youth Unemployment Rate vs Year")
plt.xticks(ind, uyear,rotation=70)
plt.show()
```

```python
#scatter plot.
plt.scatter(uyear,urate,color="b",label="unemployment rate")#green color
plt.scatter(uyear,yrate,color="g",label="youth unemployment rate")#orange color
plt.legend()
plt.ylabel("Unemployment and Youth Unemployment Rate")
plt.xlabel("Year from 1990 to 2020")
plt.title("Unemployment and Youth Unemployment Rate vs Year")
plt.xticks(uyear,rotation=70)
plt.show()
```
